```python
################# IMPORTS  ################

import os #biblioteca para manipular arquivos e pastas
from os import listdir
from os.path import isfile, join

import pandas as pd
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt


from io import StringIO
from io import BytesIO

from astropy.io import ascii, fits
from astropy.table import Table, join, hstack,vstack,QTable
from astropy.table import Table
```

```python
#################  LOAD DATA  ################
# Load samples
path = 'SAMPLES/'

file_name = 'sample_per_unit_50_training.fits'
sample_50_training = Table.read(path+file_name).to_pandas()
file_name = 'sample_per_unit_50_test.fits'
sample_50_test = Table.read(path+file_name).to_pandas()

# Load match
path = 'MATCH/'

file_name = 'match_des_sloan.fits'
match_des_sloan = Table.read(path+file_name).to_pandas()

 
path_plots = 'GRAPHICS AND ILLUSTRATIONS/'
```

```python
#################  VARIABLES  ################
Nside = 64
Npix = hp.nside2npix(Nside) # Definition of Npix - 12*(Nside)**2
```

```python
# Criando um dataframe para a plotagem do mollview 
pixels_location = pd.DataFrame()
pixels_location['indices'] = np.arange(Npix)

match_pixels_location = pd.DataFrame()
match_pixels_location['indices'] = match_des_sloan['Npix']
match_pixels_grouped = match_pixels_location.groupby(['indices']).size().reset_index(name='counts')

match_pixels_frequency = pd.merge(pixels_location, match_pixels_grouped, how = 'outer')
match_pixels_frequency = match_pixels_frequency.fillna(0) 

figure_name = 'sample_analisys_density_Objvspixel_match'

plt.figure(figsize=(10,10))
hp.mollview(match_pixels_frequency.counts.values, 
            title="Distribuição de objetos por pixel dos dados presentes no Cross Match \n entre quasares presentes no Sloan Sky Survey DR16 e objetos não classificados presentes do Dark Energy Survey ",
            unit='Objetos por pixel',
            cmap='jet',
            hold =True)
plt.savefig( path_plots + figure_name )
plt.show()

```

```python
figure_name = 'sample_analisys_density_Objvspixel_match_zoom'
plt.figure(figsize=(10,10))
hp.visufunc.gnomview(map=match_pixels_frequency.counts.values,
                     fig=None, 
                     rot=None,
                     coord=None,
                     unit='Objetos/pixel',
                     xsize=5500,
                     ysize=2500,
                     reso=1.5,
                     title='Densidade de objetos por pixel \n dos objetos pertencentes ao crossmatch ',
                     nest=False,
                     remove_dip=False,
                     remove_mono=False,
                     gal_cut=0,
                     min=None,
                     max=None,
                     flip='astro',
                     format='%.3g',
                     cbar=True,
                     cmap='jet',
                     norm=None,
                     hold=False,
                     sub=None,
                     margins=None,
                     notext=False,
                     return_projected_map=False)
plt.savefig( path_plots + figure_name )
plt.show()
```

```python
figure_name = 'sample_analisys_density_Objvspixel_training_zoom'
npix_sample_50_training_location = pd.DataFrame()
npix_sample_50_training_location['indices'] = sample_50_training['Npix']
npix_sample_50_training_objects_grouped = npix_sample_50_training_location.groupby(['indices']).size().reset_index(name='counts')

npix_sample_50_training_pixels_frequency = pd.merge(pixels_location, npix_sample_50_training_objects_grouped, how = 'outer')
npix_sample_50_training_pixels_frequency = npix_sample_50_training_pixels_frequency.fillna(0) 

plt.figure(figsize=(10,10))
hp.visufunc.gnomview(map=npix_sample_50_training_pixels_frequency.counts.values,
                     fig=None, 
                     rot=None,
                     coord=None,
                     unit='Objetos/pixel',
                     xsize=5500,
                     ysize=2500,
                     reso=1.5,
                     title='Densidade de objetos por pixel \n da amostra de treinamento',
                     nest=False,
                     remove_dip=False,
                     remove_mono=False,
                     gal_cut=0,
                     min=None,
                     max=None,
                     flip='astro',
                     format='%.3g',
                     cbar=True,
                     cmap='jet',
                     norm=None,
                     hold=False,
                     sub=None,
                     margins=None,
                     notext=False,
                     return_projected_map=False)
plt.savefig( path_plots + figure_name )
plt.show()
```
