```python
Nside = 64
Npix = hp.nside2npix(Nside) # Definição de Npix - 12*(Nside)**

# Criando um dataframe para a plotagem do mollview 
pixels_location = pd.DataFrame()
pixels_location['Npix'] = np.arange(Npix)
```

```python
sloan_quasares = Table.read('/media/new-drive/CamilaCardoso/SLOAN/SLOAN_DR16_data_quasars_pixels_location.fits').to_pandas()
quasares_per_pixel_sloan = sloan_quasares.groupby(['indices'])['indices'].count().reset_index(name="counts")
quasares_per_pixel_sloan = quasares_per_pixel_sloan.rename(columns={'indices': 'Npix'})

sloan_quasares_all_pixels = pd.merge(pixels_location, quasares_per_pixel_sloan, how = 'outer')
sloan_quasares_all_pixels = sloan_quasares_all_pixels.fillna(0)

hp.mollview(sloan_quasares_all_pixels.counts.values, title="Quantidade de quasares no SLOAN por pixel", unit=r'Quasares por pixel', cmap='jet')
```

```python
################# lê e concatena todas as tabelas que tem no diretório indicado ################
diretorio = '/media/new-drive/CamilaCardoso/DES/'
table_list = []
for diretorio, subpastas, arquivos in os.walk(diretorio):
    for arquivo in arquivos:
        df = Table.read(os.path.join(diretorio, arquivo)).to_pandas()
        table_list.append(df)
        print(os.path.join(diretorio, arquivo))
des_data = pd.concat(table_list, axis=0, ignore_index=True)
```

```python
objects_per_pixel_des = des_data.groupby(['HPIX_64'])['HPIX_64'].count().reset_index(name="counts")
objects_per_pixel_des = objects_per_pixel_des.rename(columns={'HPIX_64': 'Npix'})

des_all_pixels = pd.merge(pixels_location, objects_per_pixel_des, how = 'outer')
des_all_pixels = des_all_pixels.fillna(0)

hp.mollview(des_all_pixels.counts.values, title="Quantidade de objetos no DES por pixel", unit=r'Objetos por pixel', cmap='jet')
```
